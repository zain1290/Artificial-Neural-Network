```python
class Perceptron:

    def __init__(self, dimensions):
        self.dimensions = dimensions
        self.w = np.random.rand(dimensions)
        self.b = np.random.rand()

    def predict(self, x):
        return np.where(np.dot(x, self.w) + self.b >= 0, 1, 0)

    def train(self, X, Y, epochs=1000, learning_rate=0.05):
        for i in range(epochs):
            for x, y in zip(X, Y):
                loss = y - self.predict(x)
                self.w = [ self.w[i] + learning_rate*x[i]*loss for i in range(self.dimensions) ]
                self.b = self.b + learning_rate*loss
```

```python
def accuracy(labels, predicted):
    return np.mean(np.where(labels == predicted, 1, 0))
```

### 1.3 Neural network implementation

```python
class Linear:
    def __init__(self, in_features, out_features, init_strategy="zeros"):

        if init_strategy == "normal":
            mu, sigma = 0, 1
            self.weight = np.random.normal(mu, sigma, size=(in_features, out_features)) # (d,k)
        elif init_strategy == "xavier":
            mu, sigma = 0, np.sqrt(1/in_features)
            self.weight = np.random.normal(mu, sigma, size=(in_features, out_features))
        elif init_strategy == "he":
            mu, sigma = 0, np.sqrt(2/in_features)
            self.weight = np.random.normal(mu, sigma, size=(in_features, out_features))
        elif init_strategy == "uniform":
            self.weight = np.random.uniform(-0.1, 0.1, size=(in_features, out_features))
        else:
            self.weight = np.zeros(shape=(in_features, out_features))

        self.bias = np.zeros(out_features) # (k,)

        self.cache = None
        self.weight_grad = None
        self.bias_grad = None

    def forward(self, x):
        self.cache = x # (n,d)
        return x @ self.weight + self.bias # (n,d)(d,k) + (k,) = (n,k) + (n,k) broadcasted = (n,k)

    def backward(self, dupstream):
       # dupstream is (n,k)
        self.weight_grad = self.cache.T @ dupstream # (d,n)(n,k) = (d,k)
        self.bias_grad = np.sum(dupstream, axis=0) #(n,k) sum axis 0 = (k,)
        return dupstream @ self.weight.T # (n,k)(k,d) = (n,d)
```

```python
class Sigmoid:

    def apply(self, x):
        return 1 / (1 + np.exp(-x))

    def apply_derivative(self, x):
        return self.apply(x) * (1 - self.apply(x))
```

```python
class ReLU:

    def apply(self, x):
        return np.maximum(0, x)

    def apply_derivative(self, x):
        return np.where(x <= 0, 0, 1)
```

```python
class Activation:
    def __init__(self, function):
        self.cache = None
        self.function = function

    def forward(self, x):
        self.cache = x #(n,k)
        return self.function.apply(x) #(n,k)

    def backward(self, dupstream):
        # dupstream is (n,k)
        return dupstream * self.function.apply_derivative(self.cache) #(n,k) * C
```

```python
class BinaryCrossEntropy:
    def value(self, labels, logits):
        eps = 1e-9  # numerical stability
        y = labels.reshape(-1, 1) # (n,1)
        return -np.mean(y * np.log(logits + eps) + (1 - y) * np.log(1 - logits + eps))

    def grad(self, labels, logits):
        eps = 1e-9
        y = labels.reshape(-1, 1) # (n,1)
        return (-y / (logits + eps) + (1 - y) / (1 - logits + eps)) / y.shape[0]
```

```python
class SoftmaxCrossEntropy:

    def softmax(self, logits):
        # Stable softmax taken from the web
        shifted = logits - np.max(logits, axis=1, keepdims=True)
        exp = np.exp(shifted)
        return exp / np.sum(exp, axis=1, keepdims=True)

    def value(self, labels, logits):
        n = logits.shape[0]
        probs = self.softmax(logits)
        return -np.mean(np.log(probs[np.arange(n), labels]))

    def grad(self, labels, logits):
        n = logits.shape[0]
        probs = self.softmax(logits)
        probs[np.arange(n), labels] -= 1
        return probs / n
```

```python
class Network:
    def __init__(self, layers):
        self.layers = layers

    def forward(self, x):
        logits = x
        for layer in self.layers:
            logits = layer.forward(logits)

        return logits

    def backward(self, dupstream):
        gradient = dupstream
        for layer in reversed(self.layers):
            gradient = layer.backward(gradient)
        return gradient

    def optimizer_step(self, lr):
        for layer in self.layers:
            if isinstance(layer, Linear):
                layer.weight -= lr * layer.weight_grad
                layer.bias -= lr * layer.bias_grad
```

```python
def train(net,
          train_data, train_labels,
          val_data, val_labels,
          loss, mapToLabel, lr=0.1, epochs=1000,
          batch_size=None,
          early_stopping=False,
          patience=20,
          min_delta=0.001):

    losses = []
    accuracies = []
    n = train_data.shape[0]

    best_validation_loss = float('inf')
    epochs_without_improvement = 0
    stopping_epoch = epochs

    for epoch in range(epochs):

        indices = np.random.permutation(n)
        inputs_shuffled = train_data[indices]
        labels_shuffled = train_labels[indices]

        epoch_loss = 0

        # Normal case like before
        if batch_size is None:
            logits = net.forward(inputs_shuffled)
            loss_val = loss.value(labels_shuffled, logits)
            loss_grad = loss.grad(labels_shuffled, logits)

            net.backward(loss_grad)
            net.optimizer_step(lr)

            epoch_loss = loss_val

        else:
            for i in range(0, n, batch_size):
                X_batch = inputs_shuffled[i:i+batch_size]
                y_batch = labels_shuffled[i:i+batch_size]

                logits = net.forward(X_batch)
                loss_val = loss.value(y_batch, logits)
                loss_grad = loss.grad(y_batch, logits)

                net.backward(loss_grad)
                net.optimizer_step(lr)

                epoch_loss += loss_val * len(X_batch)

            epoch_loss /= n

        logits_full = net.forward(train_data)
        acc = accuracy(train_labels, mapToLabel(logits_full))

        losses.append(epoch_loss)
        accuracies.append(acc)

        if early_stopping:
            logits_val = net.forward(val_data)
            loss_val_validation = loss.value(val_labels, logits_val)

            if loss_val_validation < best_validation_loss - min_delta:
                best_validation_loss = loss_val_validation
                epochs_without_improvement = 0
            else:
                epochs_without_improvement += 1

            if epochs_without_improvement >= patience:
                stopping_epoch = epoch + 1
                break

    return losses, accuracies, stopping_epoch
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# FINAL MODEL — paste after all class definitions and data splits
#
# Improvements:
#   1. Adequate capacity   -- 128 and 64 neurons actually fit this problem
#   2. Momentum SGD        -- smoother updates, faster convergence
#   3. Label smoothing     -- prevents overconfidence, acts as regularisation
#   4. Proper early stop   -- patience=100, best-weights restore
#                            NO gradual decay (problem is underfitting, not overfitting)
#   5. Train on train+val  -- use 85% of labelled data for final predictions
#   6. Ensemble (11 models)-- majority vote reduces variance across random inits
# ══════════════════════════════════════════════════════════════════════════════


# Normalise unknown set with training statistics
unknown_norm = (unknown - mean) / std
np.random.seed(1)

# ─────────────────────────────────────────────────────────────────────────────
# 1. Label-smoothed Softmax Cross-Entropy  (epsilon = 0.1)
#    Replaces hard one-hot with soft targets, prevents logit explosion.
# ─────────────────────────────────────────────────────────────────────────────

class SmoothedSoftmaxCrossEntropy:

    def __init__(self, n_classes=7, epsilon=0.1):
        self.K   = n_classes
        self.eps = epsilon

    def _softmax(self, logits):
        shifted = logits - np.max(logits, axis=1, keepdims=True)
        e = np.exp(shifted)
        return e / np.sum(e, axis=1, keepdims=True)

    def _smooth(self, labels, n):
        oh = np.zeros((n, self.K))
        oh[np.arange(n), labels] = 1.0
        return oh * (1 - self.eps) + self.eps / self.K

    def value(self, labels, logits):
        probs  = self._softmax(logits)
        smooth = self._smooth(labels, logits.shape[0])
        return -np.mean(np.sum(smooth * np.log(probs + 1e-9), axis=1))

    def grad(self, labels, logits):
        probs  = self._softmax(logits)
        smooth = self._smooth(labels, logits.shape[0])
        return (probs - smooth) / logits.shape[0]


# ─────────────────────────────────────────────────────────────────────────────
# 2. Linear layer with momentum SGD
# ─────────────────────────────────────────────────────────────────────────────

class LinearMomentum(Linear):
    def __init__(self, in_features, out_features, init_strategy="he"):
        super().__init__(in_features, out_features, init_strategy)
        self.vel_w = np.zeros_like(self.weight)
        self.vel_b = np.zeros_like(self.bias)

    def step(self, lr, momentum=0.9):
        self.vel_w = momentum * self.vel_w - lr * self.weight_grad
        self.vel_b = momentum * self.vel_b - lr * self.bias_grad
        self.weight += self.vel_w
        self.bias   += self.vel_b


# ─────────────────────────────────────────────────────────────────────────────
# 3. Weight snapshot / restore
# ─────────────────────────────────────────────────────────────────────────────

def _save_weights(net):
    return [(l.weight.copy(), l.bias.copy()) if isinstance(l, Linear) else None
            for l in net.layers]

def _load_weights(net, snap):
    for layer, s in zip(net.layers, snap):
        if s is not None and isinstance(layer, Linear):
            layer.weight[:] = s[0]
            layer.bias[:]   = s[1]


# ─────────────────────────────────────────────────────────────────────────────
# 4. Training loop
#    patience=100: the previous ~76-epoch runs were stopping far too early.
#    With mini-batches and momentum, the loss curve is noisy -- you need
#    patience in the hundreds for the network to actually converge.
# ─────────────────────────────────────────────────────────────────────────────

def train_model(net, X_tr, y_tr, X_val, y_val, loss_fn,
                lr=0.01, momentum=0.9, epochs=2000,
                batch_size=64, patience=100, min_delta=1e-4):

    n             = X_tr.shape[0]
    best_val_loss = float('inf')
    best_snap     = _save_weights(net)
    bad_epochs    = 0

    for epoch in range(epochs):
        idx      = np.random.permutation(n)
        Xsh, ysh = X_tr[idx], y_tr[idx]

        for i in range(0, n, batch_size):
            Xb, yb = Xsh[i:i+batch_size], ysh[i:i+batch_size]
            net.backward(loss_fn.grad(yb, net.forward(Xb)))
            for layer in net.layers:
                if isinstance(layer, LinearMomentum):
                    layer.step(lr, momentum)
                elif isinstance(layer, Linear):
                    layer.weight -= lr * layer.weight_grad
                    layer.bias   -= lr * layer.bias_grad

        val_loss = loss_fn.value(y_val, net.forward(X_val))

        if val_loss < best_val_loss - min_delta:
            best_val_loss = val_loss
            best_snap     = _save_weights(net)
            bad_epochs    = 0
        else:
            bad_epochs += 1
            if bad_epochs >= patience:
                break

    _load_weights(net, best_snap)


# ─────────────────────────────────────────────────────────────────────────────
# 5. Build network:  10 -> 22 -> 30 -> 7
# ─────────────────────────────────────────────────────────────────────────────

def build_net():
    return Network([
        LinearMomentum(10, 22, "he"), Activation(ReLU()),
        LinearMomentum(22,  7, "xavier"),
    ])


# ─────────────────────────────────────────────────────────────────────────────
# 6. K-Fold Cross Validation Ensemble (K=5)
#
#    Trains 5 models to create the ensemble. Each model uses 80% of the 
#    combined data for training and 20% for early-stopping validation. 
#    This entirely eliminates data leakage while utilizing 100% of the data.
# ─────────────────────────────────────────────────────────────────────────────

# Combine the previously split train and val sets for Cross Validation
X_cv = np.concatenate([X_train, X_val], axis=0)
y_cv = np.concatenate([y_train, y_val], axis=0)

# Shuffle the data before splitting into folds to ensure random distribution
indices = np.random.permutation(len(X_cv))
X_cv_shuffled = X_cv[indices]
y_cv_shuffled = y_cv[indices]
K = 11
fold_size = len(X_cv) // K

loss_fn = SmoothedSoftmaxCrossEntropy(n_classes=7, epsilon=0.1)
ensemble = []

print(f"Training {K}-Fold Ensemble (10 -> 30 -> 7)...")
print("Estimated execution time: ~2-5 minutes.\n")

for i in range(K):
    # 1. Create dynamic Train/Val splits for this specific fold
    val_start = i * fold_size
    # Ensure the last fold catches any remaining elements
    val_end = (i + 1) * fold_size if i < K - 1 else len(X_cv) 
    
    # Validation chunk (20%)
    X_val_fold = X_cv_shuffled[val_start:val_end]
    y_val_fold = y_cv_shuffled[val_start:val_end]
    
    # Training chunk (80% - everything before and after the val chunk)
    X_train_fold = np.concatenate([X_cv_shuffled[:val_start], X_cv_shuffled[val_end:]], axis=0)
    y_train_fold = np.concatenate([y_cv_shuffled[:val_start], y_cv_shuffled[val_end:]], axis=0)
    
    # 2. Build and train the model for this fold
    net = build_net()
    train_model(net, X_train_fold, y_train_fold, X_val_fold, y_val_fold, loss_fn,
                lr=0.01, momentum=0.9, epochs=2000,
                batch_size=32, patience=100, min_delta=1e-4)
    
    # 3. Quick evaluation on the strictly unseen X_test to verify performance
    test_pred = np.argmax(net.forward(X_test), axis=1)
    print(f"  Fold {i+1}/{K} | test_acc = {accuracy(y_test, test_pred):.4f}")
    
    ensemble.append(net)

# ─────────────────────────────────────────────────────────────────────────────
# Ensemble Voting Logic
# ─────────────────────────────────────────────────────────────────────────────

def majority_vote(models, X):
    # Gather predictions from all models in the ensemble
    votes = np.stack([np.argmax(m.forward(X), axis=1) for m in models], axis=0)
    # Return the most frequent prediction (mode) for each sample
    return np.array([np.bincount(votes[:, j], minlength=7).argmax() for j in range(X.shape[0])])

ens_test = majority_vote(ensemble, X_test)
print(f"\nFinal Ensemble Test Accuracy: {accuracy(y_test, ens_test):.4f}")


# ─────────────────────────────────────────────────────────────────────────────
# 7. Predict unknowns and save
# ─────────────────────────────────────────────────────────────────────────────

unknown_pred = majority_vote(ensemble, unknown_norm)
predictions  = (unknown_pred + 1).reshape(1, -1).astype(int)   # 0-indexed -> 1-indexed

np.set_printoptions(suppress=True)
np.savetxt('XX_classes.txt', predictions.astype(float), delimiter=',')

print("\nSaved to 'XX_classes.txt'")
print("Preview:", ','.join(map(str, predictions[0, :10])), "...")


# ─────────────────────────────────────────────────────────────────────────────
# 8. Compare against Group_00_classes.txt
# ─────────────────────────────────────────────────────────────────────────────

try:
    their_pred = np.genfromtxt('Group_00_classes.txt', delimiter=',').flatten().astype(int)
    our_pred   = predictions.flatten()

    assert len(their_pred) == len(our_pred), \
        f"Length mismatch: theirs={len(their_pred)}, ours={len(our_pred)}"

    matches = np.sum(their_pred == our_pred)
    total   = len(our_pred)
    print(f"\nComparison with Group_00_classes.txt:")
    print(f"  Total      : {total}")
    print(f"  Matching   : {matches}  ({100*matches/total:.1f}%)")
    print(f"  Differing  : {total-matches}  ({100*(total-matches)/total:.1f}%)")

    diffs = np.where(their_pred != our_pred)[0]
    if len(diffs):
        print(f"\n  First disagreements (index | theirs | ours):")
        for idx in diffs[:10]:
            print(f"    [{idx:4d}]  {their_pred[idx]}  vs  {our_pred[idx]}")

except FileNotFoundError:
    print("\nGroup_00_classes.txt not found -- place it alongside the notebook to compare.")
```
